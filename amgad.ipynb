```python
from pyopenms import *
seq =AASequence.fromString("DFPIANANFDPGER")


print("the peptide ",str(seq),"consists of the following amino acid: ")
fullmass=0
for amino in seq:
    print(amino.getName()," : ",amino.getMonoWeight())
    fullmass=fullmass+amino.getMonoWeight()
print("full mass = ")
print(fullmass)

print("total mass = ")
print(seq.getMonoWeight())
```
